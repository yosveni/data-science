```
import numpy as np
import math
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```
## activation function step(linear)
def stepFuncao(sum):
  if sum >= 1:
    return 1
  return 
## activation function(sigmoid) for inseperably linear problem
def sigmoidFuncao(sum):
  return 1 / (1+np.exp(-sum))

## activation function(hyperbolic tanget)
def tanFuncao(sum):
  return (np.exp(sum) - np.exp(-sum) )/(np.exp(sum)+np.exp(-sum))

## activation function(relu)
def reluFuncao(sum):
  return sum if sum >= 0 else 0

## activation function(linear)
def linearFuncao(sum):
  return sum

## activation function(sofmax)
def sofmaxFuncao(x):
  exp = np.exp(x)
  return exp / exp.sum()
```

```
print('Step %s' % stepFuncao(2.1))
print('Sigmoid %s' % sigmoidFuncao(2.1))
print('Tanh %s' % tanFuncao(2.1))
print('Relu %s' % reluFuncao(2.1))
print('Linear %s' % linearFuncao(2.1))
valores = [5.0,2.0, 1.3]
print('Sofmax %s' % sofmaxFuncao(valores))
```

```
## Error variables
y_true = [1,0,1,0]
y_pred = [0.300,0.020,0.89,0.320]
```

```
## mean absolute error
print('MAE %s' % mean_absolute_error(y_true, y_pred))
print('MSE %s' % mean_squared_error(y_true, y_pred))
print('RMSE %s' % math.sqrt(mean_squared_error(y_true, y_pred)))
```
